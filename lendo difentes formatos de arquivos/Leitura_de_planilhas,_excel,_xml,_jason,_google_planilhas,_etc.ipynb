# 4. Manipulando arquivos Json

Arquivo: https://github.com/alura-cursos/Pandas/blob/main/pacientes_2.json

```python
df_json = pd.read_json('/content/pacientes_2.json')
df_json.head()
```

###**Normalizando arquivos JSON**

```python
df_normalizado = pd.json_normalize(df_json['Pacientes'])
```

```python
df_normalizado.head()
```

* Para incluir as colunas “Pesquisa” e “Ano”:

```python
#Importando a biblioteca Pandas
import pandas as pd

#Importando o módulo JSON
import json

#Lendo o arquivo json usando o módulo Python JSON
with open('pacientes_2.json','r') as f:
    dados = json.loads(f.read())

#Normalizando os dados com os parâmetros record_path e meta
pd.json_normalize(dados, record_path='Pacientes', meta=['Pesquisa', 'Ano'])

# salvando em uma variável
json_normalizado = pd.json_normalize(dados, record_path='Pacientes', meta=['Pesquisa', 'Ano'])
```

```python
json_normalizado.head()
```

* Salvar Json normalizado

```python
json_normalizado.to_json('historico_pacientes_norm.json')
```
